## Original Data Information

Initial dataset from "OtoMatch: Content-based eardrum image retrieval using deep learning" placed into three categoires:

Effusion: 179 images 
Normal: 179 images 
Tubes: 96 images 

Original dataset has some image augmentations. Some pictures were blurred, rotated, and if any words or print were on the images they were removed per article referenced below.

Reference for data: https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0232776

## Re-sorted Data Information

I re-sorted the images into five categories. I tried to place these pictures into the category which I felt was best fit. However, some of these photos were augmented and I did not know their original labels so I tried my best to evaluate which category these images would belong to. If I could not clearly see a eardrum from the image being too altered I put those images in the unclear category.

Initial dataset from "OtoMatch: Content-based eardrum image retrieval using deep learning" resorted into five categoires:

Effusion: 38 images 
Normal: 173 images 
Tubes: 96 images 
Otitis Media: 80 images
Unclear: 67 images

Images categories divided into each category with .2 of each category going to test images. Categories outline below:

#### 1) effusion (non infected fluid)
train: 31

test: 7
#### 2) normal
train: 139

test:34

#### 3) otitis media (infected fluid)
train 64

test 16

#### 4) TM tube
train: 77

test: 19

Decided to not use unclear images in training and testing of this model. When using all five categories, I got unreliable results. Better results were obtained with using just the effusion, normal, tube and otitis media categories.
